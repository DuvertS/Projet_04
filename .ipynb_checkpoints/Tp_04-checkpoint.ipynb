#### Configuration et Importations

```python
# Importations de base et configuration
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import TruncatedSVD
import warnings
warnings.filterwarnings('ignore')

# Configuration
pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
print("Bibliothèques importées et configurées")
```

### Chargement des Fichiers

```python
# Fonctions de chargement des fichiers
import glob

def load_all_datasets():
    """Charge tous les fichiers CSV disponibles"""
    datasets = {}
    available_files = glob.glob("*.csv")
    print("Fichiers détectés:", available_files)
    
    # Mapping des fichiers
    file_mapping = {
        'movies.csv': 'movies',
        'ratings.csv': 'ratings', 
        'tags.csv': 'tags',
        'links.csv': 'links',
        'genome_tags.csv': 'genome_tags',
        'genome_scores.csv': 'genome_scores'
    }
    
    for file, key in file_mapping.items():
        if file in available_files:
            datasets[key] = pd.read_csv(file)
            print(f"{file} chargé → {key}")
    
    return datasets

# Chargement des données
datasets = load_all_datasets()
print(f"\nTotal datasets chargés: {len(datasets)}")
```

### Exploration des Données

```python
# Exploration de chaque dataset
print("=== EXPLORATION DES DATASETS ===")

for name, df in datasets.items():
    print(f"\n{name.upper():<15} | Shape: {df.shape}")
    print(f"Colonnes: {list(df.columns)}")
    print(f"Valeurs manquantes:")
    print(df.isnull().sum())
    print("-" * 50)

# Focus sur le dataset movies
if 'movies' in datasets:
    movies_df = datasets['movies']
    print(f"\nAperçu de movies.csv:")
    print(movies_df.head(3))
    print(f"\nGenres uniques: {movies_df['genres'].nunique()}")
```

### Prétraitement des Films

```python
# Nettoyage et préparation des données films
def preprocess_movies_data(movies_df):
    """Nettoie et prépare les données des films"""
    df = movies_df.copy()
    
    # Nettoyage des titres
    df['clean_title'] = df['title'].apply(
        lambda x: x.split('(')[0].strip() if isinstance(x, str) else x
    )
    
    # Extraction des genres uniques
    all_genres = set()
    for genre_list in df['genres'].dropna():
        genres = genre_list.split('|')
        all_genres.update(genres)
    
    print(f"Genres uniques trouvés: {len(all_genres)}")
    print(f" {sorted(list(all_genres))[:10]}...")
    
    return df, all_genres

if 'movies' in datasets:
    movies_df, unique_genres = preprocess_movies_data(datasets['movies'])
    print(f"Films prétraités: {len(movies_df)}")
```

### Fusion des Données

```python
#  Fusion de tous les datasets
def merge_datasets(movies_df, datasets):
    """Fusionne tous les datasets avec movies"""
    enriched_df = movies_df.copy()
    
    # Fusion avec ratings
    if 'ratings' in datasets:
        ratings_df = datasets['ratings']
        movie_stats = ratings_df.groupby('movieId').agg({
            'rating': ['count', 'mean', 'std']
        }).round(2)
        movie_stats.columns = ['rating_count', 'rating_mean', 'rating_std']
        movie_stats = movie_stats.reset_index()
        
        enriched_df = enriched_df.merge(movie_stats, on='movieId', how='left')
        print(f"Ratings fusionnés → {len(enriched_df)} films")
    
    # Fusion avec tags
    if 'tags' in datasets:
        tags_df = datasets['tags']
        movie_tags = tags_df.groupby('movieId')['tag'].apply(
            lambda x: '|'.join(set(x)) if pd.notna(x).any() else ''
        ).reset_index()
        movie_tags.columns = ['movieId', 'user_tags']
        enriched_df = enriched_df.merge(movie_tags, on='movieId', how='left')
        print(f"Tags fusionnés → {len(enriched_df)} films")
    
    # Fusion avec genome
    if 'genome_scores' in datasets and 'genome_tags' in datasets:
        genome_scores = datasets['genome_scores']
        genome_tags = datasets['genome_tags']
        
        top_genome_tags = genome_scores[genome_scores['relevance'] > 0.5]
        top_genome_tags = top_genome_tags.merge(genome_tags, on='tagId')
        
        movie_genome_tags = top_genome_tags.groupby('movieId')['tag'].apply(
            lambda x: '|'.join(set(x)) if len(x) > 0 else ''
        ).reset_index()
        movie_genome_tags.columns = ['movieId', 'genome_tags']
        
        enriched_df = enriched_df.merge(movie_genome_tags, on='movieId', how='left')
        print(f"✓ Genome tags fusionnés → {len(enriched_df)} films")
    
    return enriched_df

# Fusion des données
enriched_df = merge_datasets(movies_df, datasets)
print(f"\nDataset final: {enriched_df.shape}")
print(f"Colonnes: {list(enriched_df.columns)}")
```

### Features Engineering

```python
# Création des features avancées
def create_advanced_features(enriched_df):
    """Crée des features pour le système de recommandation"""
    df = enriched_df.copy()
    
    # Combinaison des features textuelles
    df['combined_features'] = df['genres'].fillna('')
    
    if 'user_tags' in df.columns:
        df['combined_features'] += '|' + df['user_tags'].fillna('')
    
    if 'genome_tags' in df.columns:
        df['combined_features'] += '|' + df['genome_tags'].fillna('')
    
    # Score de popularité
    if 'rating_count' in df.columns and 'rating_mean' in df.columns:
        df['rating_count_norm'] = (df['rating_count'] - df['rating_count'].min()) / (df['rating_count'].max() - df['rating_count'].min())
        df['rating_mean_norm'] = (df['rating_mean'] - df['rating_mean'].min()) / (df['rating_mean'].max() - df['rating_mean'].min())
        df['popularity_score'] = 0.7 * df['rating_mean_norm'] + 0.3 * df['rating_count_norm']
        print("Score de popularité calculé")
    
    return df

# Création des features
advanced_df = create_advanced_features(enriched_df)
print(f"Features avancées créées")
print(f"Colonnes finales: {list(advanced_df.columns)}")
```

### Système de Recommandation de Base

```python
# Système de recommandation basé sur le contenu
def build_content_based_recommender(df, feature_column='genres'):
    """Construit un système de recommandation basé sur le contenu"""
    
    # Vectorisation TF-IDF
    tfidf = TfidfVectorizer(stop_words='english', tokenizer=lambda x: x.split('|'))
    tfidf_matrix = tfidf.fit_transform(df[feature_column].fillna(''))
    
    # Matrice de similarité
    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
    
    # Mapping titre-index
    indices = pd.Series(df.index, index=df['clean_title']).drop_duplicates()
    
    print(f"Système construit avec {feature_column}")
    print(f"Matrice TF-IDF: {tfidf_matrix.shape}")
    print(f"Matrice similarité: {cosine_sim.shape}")
    
    return {
        'tfidf_matrix': tfidf_matrix,
        'cosine_sim': cosine_sim,
        'indices': indices
    }

# Construction du système
recommender_system = build_content_based_recommender(advanced_df, 'combined_features')
```

### Fonctions de Recommandation

```python
# Fonctions de recommandation
def get_recommendations(title, recommender_system, df, top_n=10):
    """Retourne les recommandations pour un film donné"""
    cosine_sim = recommender_system['cosine_sim']
    indices = recommender_system['indices']
    
    try:
        idx = indices[title]
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:top_n+1]
        
        movie_indices = [i[0] for i in sim_scores]
        recommendations = df[['clean_title', 'genres']].iloc[movie_indices].copy()
        
        # Ajout des métriques si disponibles
        if 'rating_mean' in df.columns:
            recommendations['rating'] = df.iloc[movie_indices]['rating_mean'].values
        if 'rating_count' in df.columns:
            recommendations['votes'] = df.iloc[movie_indices]['rating_count'].values
        
        recommendations['similarity'] = [i[1] for i in sim_scores]
        
        return recommendations
        
    except KeyError:
        return f"Film '{title}' non trouvé."

# Test de la fonction
test_movie = "Toy Story"
recommendations = get_recommendations(test_movie, recommender_system, advanced_df)
print(f"Recommandations pour '{test_movie}':")
print(recommendations)
```

### Système de Recherche

```python
# Système de recherche de films
def search_movies(query, df):
    """Recherche des films par titre"""
    query = query.lower()
    matches = df[df['clean_title'].str.lower().str.contains(query, na=False)]
    
    if len(matches) > 0:
        results = matches[['clean_title', 'genres']].head(10)
        if 'rating_mean' in df.columns:
            results['rating'] = matches['rating_mean'].values[:10]
        return results
    else:
        return "Aucun film trouvé"

# Test de recherche
search_results = search_movies("star", advanced_df)
print("Résultats de recherche pour 'star':")
print(search_results)
```

### Interface Interactive Simple

```python
# Interface utilisateur interactive
def simple_recommender_interface():
    """Interface simple pour le système de recommandation"""
    
    print("=== SYSTÈME DE RECOMMANDATION DE FILMS ===")
    
    while True:
        print("\nOptions:")
        print("1. Rechercher un film")
        print("2. Obtenir des recommandations")
        print("3. Quitter")
        
        choice = input("\nChoisissez une option (1-3): ")
        
        if choice == '1':
            query = input("Entrez un titre ou partie de titre: ")
            results = search_movies(query, advanced_df)
            print("\nRésultats:")
            print(results)
            
        elif choice == '2':
            movie_title = input("Entrez le titre exact d'un film: ")
            print(f"\nRecherche de recommandations pour '{movie_title}'...")
            
            recommendations = get_recommendations(movie_title, recommender_system, advanced_df)
            
            if isinstance(recommendations, pd.DataFrame):
                print(f"\nTop {len(recommendations)} recommandations:")
                for i, (idx, row) in enumerate(recommendations.iterrows()):
                    print(f"{i+1}. {row['clean_title']}")
                    print(f"   Genres: {row['genres']}")
                    print(f"   Similarité: {row['similarity']:.3f}")
                    if 'rating' in row and pd.notna(row['rating']):
                        print(f"   Note: {row['rating']:.2f}")
                    print()
            else:
                print(recommendations)
                
        elif choice == '3':
            print("Au revoir!")
            break
            
        else:
            print("Option invalide. Veuillez choisir 1, 2 ou 3.")

# Décommentez pour tester l'interface
# simple_recommender_interface()
```

### Visualisations

```python
# Visualisations des données
def create_visualizations(df):
    """Crée des visualisations pour analyser les données"""
    
    # Configuration des plots
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # 1. Distribution des genres principaux
    if 'genres' in df.columns:
        all_genres = []
        for genres in df['genres'].dropna():
            all_genres.extend(genres.split('|'))
        
        genre_counts = pd.Series(all_genres).value_counts().head(15)
        axes[0,0].barh(genre_counts.index, genre_counts.values)
        axes[0,0].set_title('Top 15 des Genres')
        axes[0,0].set_xlabel('Nombre de Films')
    
    # 2. Distribution des ratings
    if 'rating_mean' in df.columns:
        ratings = df['rating_mean'].dropna()
        axes[0,1].hist(ratings, bins=20, alpha=0.7, color='skyblue')
        axes[0,1].set_title('Distribution des Notes Moyennes')
        axes[0,1].set_xlabel('Note')
        axes[0,1].set_ylabel('Nombre de Films')
    
    # 3. Nombre de votes vs note
    if 'rating_count' in df.columns and 'rating_mean' in df.columns:
        axes[1,0].scatter(df['rating_count'], df['rating_mean'], alpha=0.5)
        axes[1,0].set_xlabel('Nombre de Votes')
        axes[1,0].set_ylabel('Note Moyenne')
        axes[1,0].set_title('Votes vs Note')
        axes[1,0].set_xscale('log')
    
    # 4. Top films les mieux notés
    if 'rating_mean' in df.columns and 'rating_count' in df.columns:
        top_movies = df[df['rating_count'] > 50].nlargest(10, 'rating_mean')
        axes[1,1].barh(range(len(top_movies)), top_movies['rating_mean'])
        axes[1,1].set_yticks(range(len(top_movies)))
        axes[1,1].set_yticklabels(top_movies['clean_title'].str[:30])
        axes[1,1].set_title('Top 10 Films (min 50 votes)')
        axes[1,1].set_xlabel('Note Moyenne')
    
    plt.tight_layout()
    plt.show()

# Création des visualisations
create_visualizations(advanced_df)
```

### Évaluation du Modèle

```python
# Évaluation des performances
def evaluate_recommendation_quality(test_movies, recommender_system, df):
    """Évalue la qualité des recommandations"""
    
    results = []
    
    for movie in test_movies:
        try:
            recommendations = get_recommendations(movie, recommender_system, df)
            
            if isinstance(recommendations, pd.DataFrame):
                # Calcul de la similarité moyenne des genres
                original_genres = set(df[df['clean_title'] == movie]['genres'].iloc[0].split('|'))
                
                genre_similarities = []
                for _, row in recommendations.iterrows():
                    rec_genres = set(row['genres'].split('|'))
                    overlap = len(original_genres.intersection(rec_genres)) / len(original_genres.union(rec_genres))
                    genre_similarities.append(overlap)
                
                avg_similarity = np.mean(genre_similarities)
                results.append({'movie': movie, 'genre_similarity': avg_similarity})
                
        except Exception as e:
            print(f"Erreur avec {movie}: {e}")
            continue
    
    results_df = pd.DataFrame(results)
    
    if len(results_df) > 0:
        print("Résultats de l'évaluation:")
        print(results_df)
        print(f"\nSimilarité moyenne des genres: {results_df['genre_similarity'].mean():.3f}")
    
    return results_df

# Évaluation sur des films tests
test_movies = ["Toy Story", "Jumanji", "Heat", "Pulp Fiction", "Forrest Gump"]
evaluation_results = evaluate_recommendation_quality(test_movies, recommender_system, advanced_df)
```

### Export des Résultats

```python
# Sauvegarde des résultats
def export_results(advanced_df, evaluation_results):
    """Exporte les données et résultats"""
    
    # Sauvegarde du dataset enrichi
    advanced_df.to_csv('movies_enriched.csv', index=False)
    print("Dataset enrichi sauvegardé: movies_enriched.csv")
    
    # Sauvegarde des résultats d'évaluation
    if evaluation_results is not None:
        evaluation_results.to_csv('evaluation_results.csv', index=False)
        print("Résultats d'évaluation sauvegardés: evaluation_results.csv")
    
    # Résumé final
    print("\n=== SYNTHÈSE FINALE ===")
    print(f"Films traités: {len(advanced_df)}")
    print(f"Colonnes disponibles: {len(advanced_df.columns)}")
    
    if 'rating_mean' in advanced_df.columns:
        rated_movies = advanced_df['rating_mean'].notna().sum()
        print(f"Films notés: {rated_movies}")
    
    print("Système prêt à l'utilisation!")

# Export
export_results(advanced_df, evaluation_results)
```
